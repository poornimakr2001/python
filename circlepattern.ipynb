```python
import matplotlib.pyplot as plt
import numpy as np

# Function to create a dynamic circular pattern
def create_pattern(num_circles, radius, angle_offset):
    angles = np.linspace(0, 2 * np.pi, num_circles)
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_aspect('equal')

    for i, angle in enumerate(angles):
        x = radius * np.cos(angle + angle_offset * i)
        y = radius * np.sin(angle + angle_offset * i)
        circle = plt.Circle((x, y), radius, color='b', fill=False)
        ax.add_artist(circle)

    ax.set_xlim(-2*radius, 2*radius)
    ax.set_ylim(-2*radius, 2*radius)
    plt.show()

# Parameters
num_circles = 12
radius = 1
angle_offset = np.pi / 6  # Adjust to change the pattern

# Create the pattern
create_pattern(num_circles, radius, angle_offset)
```
