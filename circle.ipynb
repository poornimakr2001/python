```python
import matplotlib.pyplot as plt

# Set up the figure and axis
fig, ax = plt.subplots()

# Define circle properties
num_circles = 5  # Number of circles
radius = 0.1  # Radius of each circle
spacing = 0.3  # Spacing between circle centers

# Draw circles in a grid shape
for i in range(num_circles):
    for j in range(num_circles):
        circle = plt.Circle((i * spacing, j * spacing), radius, color='red', fill=True)
        ax.add_patch(circle)

# Set the aspect of the plot to be equal, so circles are not stretched
ax.set_aspect('equal')

# Set limits to ensure all circles are visible
ax.set_xlim(-0.2, num_circles * spacing)
ax.set_ylim(-0.2, num_circles * spacing)

# Hide the axes
ax.axis('off')

# Show the plot
plt.show()
```
