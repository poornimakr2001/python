```python
import cv2
from pyzbar.pyzbar import decode
import matplotlib.pyplot as plt
```

```python
def qr_code_scanner(image_path):
    # Read the image using OpenCV
    img = cv2.imread(image_path)

    # Decode the QR code
    detected_qr = decode(img)

    for qr in detected_qr:
        # Extract the bounding box to draw around the QR code
        points = qr.polygon
        (x, y, w, h) = qr.rect

        # Draw the bounding box
        cv2.rectangle(img, (x, y), (x + w, y + h), color=(255, 0, 0), thickness=2)

        # Extract and print the data in the QR code
        qr_data = qr.data.decode("utf-8")
        print(f"QR Code Data: {qr_data}")

    # Display the image with bounding box
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
