```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Set up the figure and axis
fig, ax = plt.subplots()

# Initial circle (radius=0)
circle = plt.Circle((0.5, 0.5), 0, color='blue')
ax.add_patch(circle)

# Set up the axis limits and aspect ratio
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Animation function to update the radius of the circle
def animate(frame):
    radius = frame / 100  # Increase the radius
    circle.set_radius(radius)
    return circle,

# Create animation
anim = FuncAnimation(fig, animate, frames=np.arange(0, 100), interval=50, blit=True)

# To display the animation in a notebook environment
plt.close(fig)  # Prevents duplicate display in some environments
display(HTML(anim.to_jshtml()))
```
